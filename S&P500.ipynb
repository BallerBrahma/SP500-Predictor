```python
import yfinance as yf
```

```python
sp = yf.Ticker("^GSPC")
```

```python
sp = sp.history(period="max")
```

```python
sp.plot.line(y="Close", use_index=True) #Shows the index on x axis and closing prices on y-axis
```

```python
del sp["Dividends"] #Removing Divedends and Stock Splits as they are not useful for this
del sp["Stock Splits"]
```

```python
sp["Tomorrow"] = sp["Close"].shift(-1) #Creates a new column "Tomorrow" using the pandas shift method
```

```python
sp
```

```python
sp["Target"] = (sp["Tomorrow"] > sp["Close"]).astype(int) #Returuns a int comparing the two prices
```

```python
sp
```

```python
sp = sp.loc["1990-01-01":].copy() #Only look at data after Jan 1, 1990
```

```python
sp
```

```python
from sklearn.ensemble import RandomForestClassifier #Resistant to overfit than other models. Can pick up non linear relationships
```

```python
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1) #Sets some parameters to have the best combination of efficiency, resistance, and randomness
```

```python
train = sp.iloc[:-100] #All of the rows except the last 100 in training set
test = sp.iloc[-100:] #Last 100 in testing set
predictors = ["Close", "Volume", "Open", "High", "Low"] #Predictors we are using
```

```python
model.fit(train[predictors], train["Target"]) #Training our model
```

```python
from sklearn.metrics import precision_score #Measuring performance of model
```

```python
preds = model.predict(test[predictors])
```

```python
import pandas as pd
preds = pd.Series(preds, index=test.index) #makes our prediction more readable
```

```python
preds
```

```python
precision_score(test["Target"], preds) #Gets the accuracy of our model
```

```python
comb = pd.concat([test["Target"], preds], axis=1) #Combines our test and predictions
```

```python
def predict(train, test, predictors, model): #Consolidate everything into a function
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250): #Take first 10 years, predict for 11th year and use 11 years to get 12th year and so on
    all_predictions = []
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp, model, predictors)
```

```python
predictions["Predictions"].value_counts() #Tells us how many predictions were right
```

```python
precision_score(predictions["Target"], predictions["Predictions"]) #Gives us a precision score
```

```python
predictions["Target"].value_counts() / predictions.shape[0] #Still not high enough, need more predictors
```

```python
horizons = [2,5,60,250,1000] #Gets a variety of periods
new_predictors = []

for horizon in horizons:
    rolling_averages = sp.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    sp[ratio_column] = sp["Close"] / rolling_averages["Close"] #Gets the ratio between the average close today over the different time periods
    trend_column = f"Trend_{horizon}"
    sp[trend_column] = sp.shift(1).rolling(horizon).sum()["Target"] #Looks at the past few days and takes the sum of target. 
    new_predictors += [ratio_column, trend_column]
```

```python
sp = sp.dropna()
```

```python
sp
```

```python
model = RandomForestClassifier(n_estimators=300, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model): #Define what becomes a 1 or 0
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(sp, model, new_predictors) #Only using ratios now
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
